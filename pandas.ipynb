* **_Merge function_**

```python
df1=pd.DataFrame({"calisanlar":["Ali","Veli","Aye","Fatma"],"grup":["Muhasebe","Muhendislik","Muhendislik","K"]})
df1
```

```python
df2=pd.DataFrame({"calisanlar":["Aye","Ali","Veli","Fatma"],"ilk_giris":[2010,2009,2014,2019]})
df2
```

```python
df3=pd.merge(df1,df2,on="calisanlar")
df3
```

```python
df4=pd.DataFrame({"mudur":["Caner","Mustafa","Berkcan"],"grup":["Muhasebe","Muhendislik","K"]})
df4
```

```python
df5=pd.merge(df3,df4,on="grup")
df5
```

```python
df6=pd.DataFrame({"yetenekler":["matematik","excel","kodlama","linux","excel","yonetim"],"grup":["Muhasebe","Muhasebe","Muhendislik","Muhendislik","K","K"]})
df6
```

```python
pd.merge(df5,df6,on="grup")
```

* **_.filter(func_name)_**

```python
import seaborn as sns
df=sns.load_dataset("planets")
df
```

```python
def filter_functionn(x):
    return x["mass"].std()>2
```

```python
df.groupby(["method"]).filter(filter_functionn)
```

```python
df.groupby(["method"]).std()
```
