* **_There is a difference between categories so we have to define this situation_**

```python
from pandas.api.types import CategoricalDtype
category=["Fair","Good","Very Good","Premium","Ideal"]
```

```python
df.cut=df.cut.astype(CategoricalDtype(categories=category,ordered=True))
df.cut.head()
```

```python
df["cut"].value_counts().plot.barh().set_title("Cut Variable");
```

```python
from pandas.api.types import CategoricalDtype
category1=["J","E","I","H","F","G","D"]
```

```python
df.color=df.color.astype(CategoricalDtype(categories=category1,ordered=True))
df["color"].value_counts().plot.barh().set_title("Color Variable");
```
